```python
file_name = 'properties.csv'
```

```python
data = pd.read_csv(file_name)
```

```python
data.head()
```

```python
data = data.iloc[:, :-100] # To bulk drop a massive number of columns that i deem irrelevant in the dataset
```

```python
data.head()
```

```python
columns_to_drop = ['Possession Status','Availability Starts From','Floor No','Commercial','Developer','Approved Authority Name','Price (English)','Electricity Status','Maintenance Type','Landmark','Property Uniqueness','Unit of Carpet Area','Society','Ownership Type','Parking','Facing','Data Referred From','Water Status','Luxury Flat','Covered Area','Carpet Area Unit','isPrimeLocationProperty','Tenants Preference','Location','Property']
```

```python
data = data.drop(columns=columns_to_drop)
```

```python
data.head()
```

```python
missing_values = data.isnull().sum()
print("columns with missing values")
print(missing_values[missing_values > 0])
```

```python
data_cleaned = data.dropna(how='any')
```

```python
print("Original DataFrame shape:", data.shape)
print("DataFrame shape after removing rows with any missing values:", data_cleaned.shape)
```
